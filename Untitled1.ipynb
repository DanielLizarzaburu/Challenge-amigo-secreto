```python
dia = 5
```

```python
match dia:
  case 1:
    print("Lunes")
  case 2:
    print("Martes")
  case 3:
    print("Miercoles")
  case 4:
    print("Jueves")
  case _:
    print("No coincide con niguna de las anteriores opciones")
```
