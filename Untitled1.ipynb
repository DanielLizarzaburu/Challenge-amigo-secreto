```python
# LISTAS: las listas son ordeandas, modificables y permiten valores duplicados

# indices     0         1        2         3
frutas = ["Manzana", "Papaya", "Pera", "Manzana"]
print(frutas)
print(type(frutas))

frutas[1] = "Platano"
print(frutas)
print(frutas[1])

print("------------")

lista = ["Buru", 11, True]
print(lista)
print(type(lista))

print("----------")

print(len(lista)) # la funcion len, nos dice cuantos datos tiene dicho conjunto
print(len(frutas))

print(frutas[1:3])
print(frutas[:3])
print(frutas[2:])

if "Manzana" in frutas:
  print("Si hay Manzana en la lista")
```

```python
# indice       0          1         2
vehiculos = ["Auto", "Bicicleta", "Moto"]

# metodos
# append, (agrega un elemento al final de la lista)
vehiculos.append("Escuter")
print(vehiculos)
# insert, agrega un elemento en la posicion que indiquemos
vehiculos.insert(1, "Minivan")
print(vehiculos)
# extend, agrega elementos al final de la lista
vehiculos.extend(["Camion", "Tren"])
print(vehiculos)
#remove, elemina un elemento
vehiculos.remove("Moto")
print(vehiculos)
# pop, elimina un elemento especificando su indice
vehiculos.pop(2)
print(vehiculos)
#sort, ordena la lista de forma alfabetica
vehiculos.sort()
print(vehiculos)
# reverse, invierte el orden de la lista
vehiculos.reverse()
print(vehiculos)
```

```python
frutas = {"Manzana", "Platano", "Cereza", "Platano"}
print(frutas)
print(type(frutas))

# agregar
# add, agrega un elemento al conjunto
frutas.add("Pera")
print(frutas)

# update, agrega un conjunto de elementos, se pueden agrenar listas, tuplas, conjuntos
frutasAdicionales = {"Naranja", "Melon", "Fresa"}
frutas.update(frutasAdicionales)
print(frutas)

# remove, elemina un elemento del conjunto, si o si tiene que aparecer en el conjunto o sino da error
frutas.remove("Melon")
print(frutas)

# discard, elemina un elemento del conjunto, asi no aparezca en el conjunto
frutas.discard("Papaya")
print(frutas)

# pop, elimina uno al alzar
frutas.pop()
print(frutas)

# clear, elimina todos los elementos del conjunto
frutas.clear()
print(frutas)

# del, elimina el conjunto
del frutas

```

```python
# union, interseccion y reta de conjuntos
a = {2, 3, 6,}
b = {4, 5, 6}

c = a.union(b)
print(c)

d = a.intersection(b)
print(d)

e = a.difference(b)
print(e)
```

```python
# diccionario python, coleccion de pares clave calor(ordenado a partir de python3.7)
# Un diccionario es una colección de pares clave-valor, lo que lo hace muy útil para almacenar datos estructurados.

auto = {
    "marca": "Renault",
    "modelo": "Clio",
    "año": 2023
}
print(auto)
print(type(auto))
print(auto["marca"]) # imprime el dato de la marca
print(auto.get("modelo")) # otra forma de imprimir el dato
print(auto.keys()) # keys son las claves, por ejemplo una clave es "marca"
print(auto.values()) # values imrpime los valores de las claves, un ejemplo de un valor es "Renault"
print(auto.items()) # imprime todos los pares clave-valor

if "modelo" in auto:
  print("Si hay un modelo en el diccionario")
```

```python
# diccionarios
auto = {
    "marca": "Renault",
    "modelo": "Clio",
    "año": 2023
     }

auto["año"] = 2020  # modifica el año
print(auto)

auto["color"] = "verde" # agrega un clave-valor
print(auto)

auto.update({"año": 2022, "puertas": 4}) # modifica el año y agrega un clave-valor nuevo
print(auto)

auto.pop("año") # elimina el año
print(auto)

auto.popitem() # elimina el ultimo elemento
print(auto)

auto.clear() # limpia el diccionario, y solo da el resultado de las { }
print(auto)
```

```python
# diccionarios
auto = {
    "marca": "Renault",
    "modelo": "Clio",
    "año": 2023,
    "color": "Verde",
    "puertas": 4
     }

for k in auto: # keys
  print(k)
print("--------")

for v in auto.values(): # values
  print(v)
print("--------")

for k, v in auto.items(): # items
  print(k, v)
```

```python
# OTRO CASO LAMBDA

def mifuncion(n):
  return lambda a : a * n

doble = mifuncion(2)     # en vez de doble también usan la palabra duplicador
triple = mifuncion(3)     # también usan la palabra triplicador
quintuple = mifuncion(5)  # también usan la palabra quintuplicar

print(doble(11))
print(triple(11))
print(quintuple(11))
```

```python
# Lectura y escritura de texto en Python

try:
# el error es porque el archivo no existe, entonces saltamos el error con try
  f = open("archivo.txt", "r")
  print(f.readline())
  f.close() #cierra la lectura del archivo
except FileNotFoundError:
  print("El archivo no existe")
```

```python
# otra forma donde no es necesario cerrar el archivo con close, para evitar muchas lineas de código
try:
  with open("archivo.txt", "r") as f:
    print(f.readline())
except FileNotFoundError:
  print("El archivo no existe")
```
